```python
# Write a programe arithmatic operator using switch statement
x, y = 10, 20
option = input("Enter your choice option :")
match option :
    case '+' :
      print("sum of values", x + y)
    case '-' :
      print("subtraction of value", x - y)
    case '*' :
      print("Multiplication of value", x * y)
    case '/' :
      print("Division of value", x / y)
    case '%' :
      print("Modulus of value", x % y)
    case '_' :
      print("Invalide operator")
    case '!=' :
      print("Not equal of value", x != y)
```
