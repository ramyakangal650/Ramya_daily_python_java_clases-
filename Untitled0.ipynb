```python
class BCA_1st_semister :
    def student(self,a,b=0,c=0,d=0):
      return a+b+c+d

BCA = BCA_1st_semister()
print(BCA.student(5))
print(BCA.student(10,7,8))
print(BCA.student(-10,4,9,-5))
print(BCA.student(99,0,4,-12))
```

```python
class Animal :
  def sound(self):
    print("Animal makes a sound")

class Dog(Animal):
  def sound(self):
    print("Bow Bow Bow Bow")

class Cat(Animal) :
  def sound(self):
    print("Meow Meow Meow Meow")

class Tiger(Animal):
  def sound(self):
    print("Roar Roar Roar Roar")

class Snake(Animal):
  def sound(self):
    print("Buuuussssssssssssssss")

class Cow(Animal):
  def sound(self):
    print("Ambbbbbbbaaaaaaa")

obj1 = Animal()
obj2 = Dog()
obj3 = Cat()
obj4 = Tiger()
obj5 = Snake()
obj6 = Cow()

obj1.sound()
obj2.sound()
obj3.sound()
obj4.sound()
obj5.sound()
obj6.sound()
```

```python
class Book :
  def __init__(self,page):
    self.page = page
  def __add__(self,other):
    return self.page + other.page

obj1 = Book(200)
obj2 = Book(300)
print("Total pages is",obj1 + obj2)
```

```python
# Abstraction Examples
from abc import ABC, abstractmethod
class Game(ABC):

  # normal funcion / method
    def Free_fire(self):
        print("I'm doing work")

  # Abstract Method / function
    def BGMI(self):
        pass

    # Normal function / method
    def Ludo_King(self):
        print("winner winner chicken dinner")

class Temple_Run (Game):

  # BGMI fnction implemented by sub class
  def BGMI(self):
    print("I got Royal pass Daimond")

  # Normal function
  def chess(self):
    print("I'm a district player")

obj = Temple_Run()
obj.Free_fire()
obj.BGMI()
obj.Ludo_King()
obj.chess()
```
